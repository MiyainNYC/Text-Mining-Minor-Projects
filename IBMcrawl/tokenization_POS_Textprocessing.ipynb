```python
import json
import pandas as pd
import json
import nltk
import csv
from nltk import FreqDist
from nltk.corpus import stopwords
```

```python
stopwords = nltk.corpus.stopwords.words('english')
operators = ['ibm', 'cloud','computing']
porter =nltk.PorterStemmer()
```

```python
year_to_look = str(raw_input("Which year you are gonna explore?\n"))
```

    Open the file and dump into json

```python
with open(year_to_look+'.json','rU') as f:
    raw = f.read()
    raw_encode = json.dumps(raw)
```

    How many articles about IBM cloud in a paticular year?

```python
with open(year_to_look+'.json') as json_data:
    d = json.load(json_data)
raw_pd = pd.DataFrame(d)
print "There are %r articles in %r" %(len(raw_pd),int(year_to_look))
```

Tokenization

```python
tokens = nltk.word_tokenize(raw_encode)
text = nltk.Text(tokens)
```

```python
Num_words_in_total = len(text)
print "There were %d words in %d year talking about IBM cloud." % (len(text),int(year_to_look))
```

Frequent Dist for every word after removing stopwords but not stemming

```python
content_lower = [w.lower() for w in text]
content = [w for w in content_lower if w not in stopwords if w not in operators if w.isalpha()]
frdist_for_each_word = FreqDist(content)
rank_each_word = sorted(frdist_for_each_word.items(), key = lambda x:x[1],reverse = True)
rank_each_word[0]
```

```python
rank_pd = pd.DataFrame(rank_each_word[:50],columns = ("word","count"),index = range(1,51))
print "Here are the 50 most frequent words written in year %d and their frequency: \n\n%s" % (int(year_to_look),
                                                                         rank_pd)
```

```python
filename_1 = "most_50_common_words_"+year_to_look+".csv"
rank_pd.to_csv(filename_1, index=False)
```

```python
word_occur_once= sorted(frdist_for_each_word.hapaxes())

print "Here are %d words appear only once in articles written in %d:\n%s" % (len(word_occur_once),
                                                                             int(year_to_look), 
                                                                             "\n".join([x for x in word_occur_once])
                                                                            )
```

POS

```python
tagged = nltk.pos_tag(content)
word_NNP= [(word,tag) for (word,tag) in tagged if tag.startswith('NNP')]
entities = nltk.chunk.ne_chunk(tagged)
word_NNP
```

Frequenct distribution for NNP 

```python
word_NNP_fd = FreqDist(word_NNP)
rank_NNP = sorted(word_NNP_fd.items(),key = lambda x:x[1],reverse = True)
rank_NNP_pd = pd.DataFrame(rank_NNP,columns = ('word','count'))
print "Here are NNP words in articles written in %d:\n\n %s" % (int(year_to_look),
                                                               rank_NNP_pd)
```

```python
filename = "word_NNP_fd_"+year_to_look+".csv"
rank_NNP_pd.to_csv(filename,index = False)
```
